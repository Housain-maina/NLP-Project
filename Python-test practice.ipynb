Question 8: MaxCounter Codility Test Explained – Full Explanation Tutorial for Max Counters 

```
def solution(N, A):
    
    counters = [0] * N 
    max_1 = 0
    max_2 = 0

    for x in A:
        if 1 <= x <= N:
            if max_1 > counters[x-1]:
                counters[x-1] = max_1
            counters[x-1] += 1

            if max_2 < counters[x-1]:
                max_2 = counters[x-1]
        
        else:
            max_1 = max_2
    
    counters = [max(max_1, i) for i in counters]

    return counters



A = [3, 4, 4, 6, 1, 4, 4]
N = 5

print(solution(N, A))
```

Question 16: Disc Intersections Codility 

```
from bisect import bisect_right

def solution(A):

    start = [i-j for i, j in enumerate(A)]
    start.sort()
    pairs = 0

    for i in range(0, len(A)):
        end = i + A[i]
        count = bisect_right(start, end)
        count_1 = count-1
        count_2 = count_1 - i
        pairs += count_2

    return pairs

A = [-3, 1, 2, 4, 3, 3, -4, 7, 9, 11, 12]

print(solution(A))
```
