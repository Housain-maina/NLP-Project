Efficient algorithm to find the largest run of zeros in a binary string? 

```
def longRunZeros(s):
    big = 0
    curr = 0
    for c in s:
        if c == '0':
            curr += 1
        else:
            if curr > big:
                big = curr
            curr = 0
    if curr > big:
        big = curr
    return big

longRunZeros('00100110100 0')
```

```
# Python3 code to determine Length of  
# longest consecutive zeroes in the 
# binary representation of a number.  
def maxZeros(N): 
  
    # variable to store the length  
    # of longest consecutive 0's 
    maxm = -1
  
    # to temporary store the  
    # consecutive 0's 
    cnt = 0
    while(N): 
        if(not(N & 1)): 
            cnt += 1
            N >>= 1
            maxm = max(maxm,cnt) 
        else: 
            maxm = max(maxm,cnt) 
            cnt = 0
            N >>= 1
  
    return maxm 
  
# Driver Code 
N = 14
print(maxZeros(N)) 
```

Question 2. Count the numbers divisible by ‘M’ in a given range


A and B are two numbers which define a range, where $A <= B$. Find the total numbers in the given range $[A … B]$ divisible by 'M'

```
# Program to count the numbers  
# divisible by M in a given range 
  
def countDivisibles(A, B, M): 
      
    # Variable to store the counter 
    counter = 0; 
  
    # Running a loop from A to B 
    # and check if a number is  
    # divisible by M. 
    for i in range(A, B): 
        if (i % M == 0): 
            counter = counter + 1
  
    return counter 
  
# Driver code 
# A and B define the range, 
# M is the dividend 
A = 30
B = 100
M = 30
  
print(len(range(A,B)))

# Printing the result 
print(countDivisibles(A, B, M)) 
```

```
def countDivisibles(A, B, M): 
      
    # Variable to store the counter 
    counter = 0; 
  
    # Running a loop from A to B 
    # and check if a number is  
    # divisible by M. 
    for i in set(range(A, B+1)): 
        if (i % M == 0): 
            counter = counter + 1
  
    return counter 
  
# Driver code 
# A and B define the range, 
# M is the dividend 
A = 30
B = 100
M = 30

print(len(set(range(A, B+1))))
# Printing the result 
print(countDivisibles(A, B, M)) 
```

Question 3: cyclic rotation challenge

```
def cyclicRotation(A, K):
    N = len(A)
    B = [0]*N
    print(B)

    for i in range(N):
        B[i] = A[(i - K) % N]
    return B

print(cyclicRotation([2,3,4,5], 2))
```

```
def cyclicRotation(A, K):
    if len(A) == 0:
        return A
    K = K % len(A)
    return A[-K:] + A[:-K]

print(cyclicRotation([2,3,4,5], 2))
```

Question 8: MaxCounter Codility Test Explained – Full Explanation Tutorial for Max Counters 

```
def solution(N, A):
    
    counters = [0] * N 
    max_1 = 0
    max_2 = 0

    for x in A:
        if 1 <= x <= N:
            if max_1 > counters[x-1]:
                counters[x-1] = max_1
            counters[x-1] += 1

            if max_2 < counters[x-1]:
                max_2 = counters[x-1]
        
        else:
            max_1 = max_2
    
    counters = [max(max_1, i) for i in counters]

    return counters



A = [3, 4, 4, 6, 1, 4, 4]
N = 5

print(solution(N, A))
```

Question 15: Max Product of Three P, Q, R




```
import heapq

def solution(A):
    heapq.heapify(A)

    large = heapq.nlargest(3, A)
    large1, large2, large3 = large[0], large[1], large[2]

    small = heapq.nsmallest(2, A)
    small1, small2 = small[0], small[2]

    return max(large1 * large2 * large3, small1 * small2 * large1)

A = [-3, 1, 2, 4, 3, 3, -4, 7, 9, 11, 12]

print(solution(A))
```

Question 16: Disc Intersections Codility 

```
from bisect import bisect_right

def solution(A):

    start = [i-j for i, j in enumerate(A)]
    start.sort()
    pairs = 0

    for i in range(0, len(A)):
        end = i + A[i]
        count = bisect_right(start, end)
        count_1 = count-1
        count_2 = count_1 - i
        pairs += count_2

    return pairs

A = [-3, 1, 2, 4, 3, 3, -4, 7, 9, 11, 12]

print(solution(A))
```
